CSS can be defined 
1. inline
   <p style='color: red; font-size: 20px;'>I'm learning to code!</p>
3. internal stylesheet
   <head>
  <style>
    p {
      color: red;
      font-size: 20px;
    }
  </style>
</head>

3. separate file, which must be linked in the HTML head
    <link href='https://www.codecademy.com/stylesheets/style.css' rel='stylesheet'>
    or
    <link href='./style.css' rel='stylesheet'>


### CSS selectors
1. **type** selector matches the type of the element in the HTML document. It is also called element or tag name selector
    h1 {
   color: maroon;
   }

2. **universal** selector selects all elemets of any type
   * {
     border: 1px solid red;
     }
     
3. **class** selector
   the most common way to select elements.
   In HTML tags you can define a class attribute and in CSS you can select all the elements
   with the same class -> .title {color:teal;} . Class selector overwrites the type selector

You can give multiple class names in the HTML tags separated by space.
This is useful if you have multiple rulesets in thee CSS and you want to apply more of them to the same element:
CSS classes:
.green {color: green;}
.bold {font-weight: bold;}

HTML
<h1 class="bold green">...</h1>

4. **select by ID**
   if we want to style a single targeted element we can use its ID:
   #html_tag_id { ... }
   the ID selector overwrites everything, so we should use it only onm elements which must remain the same everywhere

5. **Attribute selector**
   either with brackets all the attributes:
   [href]{color: magenta;}
   or more granularly with this syntax: type[attribute*=value]
   img[src*='winter'] {height: 50px;} - will target the img with winter in its source
   Advantage: no new HTML markup (like a class or id) needed to be added

6. **pseudo class**
   :focus, :visited, :disabled, :active, :hover

7. **Specificity**
   the browser decides which style to use based on how specific a given selector is.
   type is the weakest, then class and ID overwrites everything. Chaining and combining selectors increases their specificity.

8. **Chaining** is the process when we use a typ and a class selector combined
   h2.destination {font-family: Tahoma;}

9. **Descendant Combinator**
    CSS also supports selecting elements that are nested within other HTML elements, also known as descendants.
   .main-list li { ... } - selects the li tags within the ul tag with class main-list

10. **Multiple selectors**
    If the same style applies to more selectors you can list them with commas:
    h1, .menu {font-family: Georgia;}
