### Create a table

Before we can put data in the table we have to create its structure
TABLE tag gives theframe
TR tags are the rows in the table
within the rows the data must be added in each cell with TD
The headers can be myde by the table heading TH tags, they must be the children or a TR
TH tags has an attribute indicating whether they are for a column or a row (scope = "col" or scope = "row")
In older HTML versions the border was added by the border attribute in the table tag
Nowadays we use css for formatting tables too
If you want to allow a TD element to span more columns you can use the colspan attribute (<td colspan="2" ... the integer >=1)
Also rows can span into multiple rows with <td rowspan="2" ... integer>=1

If the table grows too long/big we can separate the body between TBODY tags and
put the heading into THEAD tags and
put the footer into TFOOT tags
