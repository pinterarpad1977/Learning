### HTML Anatomy

An element is an opening and a closing tag together with the content between them.


### The Body
<body> content </body>
- only content inside the body tags will be displayed on the page
- every element nested into another element is a child element and the element it is nested in is the parent
- this relationship between the elements is called hierarchy
- hierarchy is important because the children can inherit behaviour and styling from theis parents in CSS

### Attributes
Attributes are content added to the opening tag of an element and can be used in several different ways, from providing information to changing styling. Attributes are made up of the following two parts:

- The name of the attribute
- The value of the attribute

One commonly used attribute is the id. We can use the id attribute to specify different content (such as <div>s) and is really helpful when you use an element more than once. 

<body>
  <h1>The Brown Bear</h1>
  <div id="introduction">
    <h2>About Brown Bears</h2>
    <h3>Species</h3>
    <h3>Features</h3>
  </div>
  <div id='habitat'>
    <h2>Habitat</h2>
    <h3>Countries with Large Brown Bear Populations</h3>
    <h3>Countries with Small Brown Bear Populations</h3>
  </div>
  <div id='media'>
    <h2>Media</h2>
  </div>
</body>

### Displaying text
You can use 
- paragraphs (p-tags) or
- span tag -> containes a short piece of thext or html, used to separate small pieces of content on the same line as the text
- span is included in the p-tag

<body>
  <h1>The Brown Bear</h1>
  <div id="introduction">
    <h2>About Brown Bears</h2>
      <p>The brown bear (Ursus arctos) is native to parts of northern Eurasia and North America. Its conservation status is currently <span>Least Concern</span>. There are many subspecies within the brown bear species, including the Atlas bear and the Himalayan brown bear.</p>
    <h3>Species</h3>
    <h3>Features</h3>
      <p>Brown bears are not always completely brown. Some can be reddish or yellowish. They have very large, curved claws and huge paws. Male brown bears are often 30% larger than female brown bears. They can range from 5 feet to 9 feet from head to toe.</p>
  </div>
  <div id="habitat">
    <h2>Habitat</h2>
    <h3>Countries with Large Brown Bear Populations</h3>
    <h3>Countries with Small Brown Bear Populations</h3>
      <p>Some countries with smaller brown bear populations include Armenia, Belarus, Bulgaria, China, Finland, France, Greece, India, Japan, Nepal, Poland, Romania, Slovenia, Turkmenistan, and Uzbekistan.</p>
  </div>
  <div id= "media">
    <h2>Media</h2>
  </div>
</body>

### Images (img)
img tag is self closing
self closing tags may omit or include /, both will render properly
required attribute is src which must be the URL(Uniform Resource Locator) of the image
the URL can be web or local address too

<img src="https://content.codecademy.com/courses/web-101/web101-image_brownbear.jpg">


### Image Alts
the content within the alt attribute will be used by screen readers
the alt  must be within the img tag with a text depicting the image
if the image doesn't load the alt text will be visible if the user moves the mouse over the image area
it plays a role also in the SEO (Search Engine Optimization), because search engines cannot see the pictures

<img src="https://content.codecademy.com/courses/web-101/web101-image_brownbear.jpg" alt="Big brown bear is looking into the camera in a green forest" />

### Video (video)
like the img tag it needs a source attribute
Additionally it needs the width and the height attributes and
we can show basic controls with the control attribute
an additional text can be added between the video tags which appears only if the browser is unable to load the video
